```python
# import pandas
import pandas as pd

df = pd.read_csv("FAISS_text.csv")

df
```

```python
# converting the text to vectors using embeddings

from sentence_transformers import SentenceTransformer
```

```python
encoder = SentenceTransformer("all-mpnet-base-v2")
```

```python
vectors = encoder.encode(df.text)
vectors
vectors.shape
```

```python
dim =vectors.shape[1]
```

```python
import faiss
# use eucledian distancing to do the indexing
```

```python
index=faiss.IndexFlatL2(dim)
# in simple terms i had my dataframes and later each individual text is converted into vector embeddings
#then i told the program to add the vectors into the free space i have created
index
```

```python
index.add(vectors)
```

```python
# now as i said above i have added the vectors into the index thats whi earlier i have stored the dimention of the vectors
#so i wont have a problem here
#in the following code files i am using eucledian distance of the vectors fron thew required and use that distance to output my required vector
```

```python
# i am using an random query
search_query = "Looking for places to visit with my girlfriend"
query_2="Looking for job?"
```

```python
vec=encoder.encode(search_query)
vec2=encoder.encode(query_2)
vec.shape# i want this vector to be a 768,1 dimention vector so that i can use eucledian distance so for that i need numpy
```

```python
import numpy as np
sup_vec = np.array(vec).reshape(1,-1)
sup_vec_2=np.array(vec2).reshape(1,-1)
sup_vec.shape
sup_vec_2.shape
```

```python
# now i am gonna search this vector
index.search(sup_vec,k=2)
index.search(sup_vec_2,k=2)
```

```python
distances_1, indices_1 = index.search(sup_vec, k=2)  # Search for 2 nearest neighbors for query 1
distances_2, indices_2 = index.search(sup_vec_2, k=2)  # Search for 2 nearest neighbors for query 2
```

```python
nearest_rows_1 = df.loc[indices_1[0]]
nearest_rows_2 = df.loc[indices_2[0]]
print(nearest_rows_2)
```
